# **Import Dependencies**

```python
import pandas as pd
import seaborn as sns
```

# **Load Data**

Sumber data: https://www.hackerrank.com/challenges/battery/problem

```python
dataset_link = 'https://raw.githubusercontent.com/dinanabila/hackerrank-scratchwork/refs/heads/main/Artificial%20Intelligence/Laptop%20Battery%20Life/trainingdata.csv'
df = pd.read_csv(dataset_link, header=None, names=['charging_duration','lasted_duration'])
```
